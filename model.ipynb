### Modeling

```python
# --- 2. Pembersihan dan Vektorisasi Teks ---
# Menggabungkan judul dan deskripsi untuk representasi teks yang lebih kaya
# courses_df['teks_gabungan'] = courses_df['judul'] + " " + courses_df['deskripsi']

# Pilihan Vektorisasi: TF-IDF atau SentenceTransformer
# Untuk contoh ini, kita akan menggunakan kedua metode secara terpisah untuk demonstrasi

# --- Pendekatan 2a: Menggunakan TF-IDF ---
print("--- Menggunakan TF-IDF ---")
tfidf_vectorizer = TfidfVectorizer(stop_words='english')
# Fitur TF-IDF dari teks gabungan
tfidf_matrix = tfidf_vectorizer.fit_transform(courses_classentral_df['cleaned_text'])

def recommend_courses_tfidf(query, tfidf_matrix, courses_df, vectorizer, top_n=3):
    """
    Merekomendasikan kursus menggunakan TF-IDF dan cosine similarity.
    """
    # Vektorisasi query pekerjaan
    query_vector = vectorizer.transform([query])

    # Hitung cosine similarity antara query dan semua deskripsi kursus
    cosine_similarities = cosine_similarity(query_vector, tfidf_matrix).flatten()

    # Dapatkan indeks kursus dengan similaritas tertinggi
    related_course_indices = cosine_similarities.argsort()[:-top_n-1:-1]

    # Tampilkan hasil rekomendasi
    print(f"\nKursus yang direkomendasikan untuk '{query}':")
    count = 0
    for i in related_course_indices:
        if cosine_similarities[i] >= 0.5 and count < top_n:
            print(f"- Judul: {courses_df['Title'].iloc[i]}")
            print(f"  Platform: {courses_df['Provider'].iloc[i]}")
            print(f"  Skor Similarity: {cosine_similarities[i]:.4f}")
            print("-" * 20)
            count += 1
        elif count == top_n:
            break
    if count == 0:
        print("No recommendations with a similarity score of 0.5 or higher.")

# Contoh penggunaan TF-IDF
recommend_courses_tfidf("data scientist", tfidf_matrix, courses_classentral_df, tfidf_vectorizer)
```
